```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import skfda
from skfda.representation.basis import BSplineBasis
from skfda.preprocessing.smoothing import BasisSmoother
from skfda.preprocessing.dim_reduction.feature_extraction import FPCA


# ----------------------------
# 2. Group by 'chunk' to extract individual curves and metadata.
#    We assume each CSV (chunk) has rows with the same "Filtration Value" sequence.
# ----------------------------
curve_list = []     # will store y-values of the curve for each chunk
meta_list = []      # will store a dict of metadata for each curve

# Group by unique chunk; each group corresponds to one curve.
grouped = all_data.groupby("chunk")
for chunk_id, group in grouped:
    # Sort by "Filtration Value" to ensure the curve is in order
    group_sorted = group.sort_values("Filtration Value")
    x = group_sorted["Filtration Value"].values
    y = group_sorted["Raw Euler Characteristic"].values

    # Store the raw y values
    curve_list.append((x, y))
    
    # Get metadata (assuming these are constant within the chunk)
    row = group_sorted.iloc[0]
    meta_list.append({
        "chunk": chunk_id,
        "sample": row["sample"],
        "section": row["section"]
    })

# ----------------------------
# 3. Interpolate all curves to a common grid
# ----------------------------
# Define a common grid. Here we use the min and max from the first curve.
# (Adjust if needed or compute a global min/max from all curves.)
x0, _ = curve_list[0]
common_grid = np.linspace(np.min(x0), np.max(x0), len(x0))

# Interpolate each curve onto the common grid
curves_interp = []
for x, y in curve_list:
    # Linear interpolation; assumes x is sorted
    y_interp = np.interp(common_grid, x, y)
    curves_interp.append(y_interp)

# Convert list of curves to a 2D numpy array (n_curves x n_grid_points)
data_matrix = np.array(curves_interp)

# ----------------------------
# 4. Create a functional data object and fit B-spline basis functions
# ----------------------------
# Create an FDataGrid object from the data_matrix and the common grid.
fd_grid = skfda.FDataGrid(data_matrix, grid_points=common_grid)

# Define a B-spline basis.
# Adjust n_basis (e.g. 10) as needed for the smoothness you desire.
bspline_basis = BSplineBasis(n_basis=10, domain_range=(common_grid[0], common_grid[-1]))

# Create a smoother to fit the B-spline basis to the raw data.
smoother = BasisSmoother(basis=bspline_basis)
fd_basis = smoother.fit_transform(fd_grid)

# ----------------------------
# 5. Perform Functional PCA
# ----------------------------
# Here we extract, for example, 2 principal components.
fpca = FPCA(n_components=2)
fpca.fit(fd_basis)
scores = fpca.transform(fd_basis)  # shape: (n_curves, 2)

# Create a DataFrame for the scores, merging in metadata.
meta_df = pd.DataFrame(meta_list)
scores_df = pd.DataFrame(scores, columns=["PC1", "PC2"])
meta_df = pd.concat([meta_df, scores_df], axis=1)

# ----------------------------
# 6. Plot the FPCA scores colored by sample (or section)
# ----------------------------
plt.figure(figsize=(8, 6))
for sample in meta_df["sample"].unique():
    subset = meta_df[meta_df["sample"] == sample]
    plt.scatter(subset["PC1"], subset["PC2"], label=sample)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Functional PCA of B-spline fitted curves")
plt.legend(title="Sample")
plt.show()

# Optionally, if you want to inspect the fitted curves:
plt.figure(figsize=(10, 6))
for i, meta in meta_df.iterrows():
    plt.plot(common_grid, data_matrix[i, :], label=meta["sample"], alpha=0.5)
plt.xlabel("Filtration Value")
plt.ylabel("Raw Euler Characteristic")
plt.title("Raw curves (one per chunk) colored by sample")
plt.show()
```
